```python
import optuna
from optuna.trial import create_trial

# Source and destination databases
SOURCE_DB = "sqlite:///Data/optuna/lunarlander/add_novelty/diff_fixed.db"
DEST_DB = "sqlite:///Data/optuna/lunarlander/add_novelty/diff.db"

STUDY_NAME_S = "diff_add_novelty_exp_fixed"
STUDY_NAME_T = "diff_add_novelty_exp_fixed"


# Load source study
src = optuna.load_study(
    study_name=STUDY_NAME_S,
    storage=SOURCE_DB,
)

# Create destination study
dst = optuna.create_study(
    study_name=STUDY_NAME_T,
    storage=DEST_DB,
    direction=src.direction,  # or directions=src.directions for multi-objective
    load_if_exists=True,
)

# Copy trials
for trial in src.trials:
    copied = create_trial(
        params=trial.params,
        distributions=trial.distributions,
        value=trial.value,
        state=trial.state,
        user_attrs=trial.user_attrs,
        system_attrs=trial.system_attrs,
        intermediate_values=trial.intermediate_values,
    )
    dst.add_trial(copied)

print(f"Copied {len(src.trials)} trials")
```
